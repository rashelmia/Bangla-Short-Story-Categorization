```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from bs4 import BeautifulSoup 
import os 
os.chdir("F:\DIU\Thesis")
driver = webdriver.Firefox()
```

### Insert the website link

```python
url = 'https://bengali.pratilipi.com/premkahini'
driver.get(url)
```

```python
novels = driver.find_elements_by_class_name('summary-card')
```

### Extract the Story Title, Author Name and Links of the story

```python
novels = driver.find_elements_by_class_name('summary-card')
num_of_novel = 0
list_of_link = []
title_of_novel  = []
author_name = []
author = []
for novel in novels:
    title = novel.find_element_by_class_name("book-title").text
    print(title)
    title_of_novel.append(title)
    
    author = novel.find_element_by_class_name('author-name').text
    print(author)
    author_name.append(author)
    
    #summary= novel.find_element_by_class_name('summary').text
    
    elems = novel.find_elements_by_css_selector(".summary [href]")
    links = [elem.get_attribute('href') for elem in elems]
    #print(links)
    for link in links:
        #link=  link.get_attribute('href') 
        list_of_link.append(link)
        #print(list_of_link_for_text)
    num_of_novel = num_of_novel+1
print('Number of Novel:',num_of_novel)
dataframe = pd.DataFrame({'Title':title_of_novel,'Author':author_name})
```

### Get the text from those links

```python
text = ''
details ={}
all_text = []
num_of_chapter = 0
for i in range(0,num_of_novel):
    url = list_of_link[i]
    driver.get(url)
    time.sleep(1)

    link = driver.find_element_by_class_name("read-btn")
    link.click()
    time.sleep(1)
    
    #text from the each chapter
    containers  = driver.find_elements_by_class_name('book-content [data-suggestions]')
    for container in containers:
        history = container.get_attribute('data-suggestions')
        text = text+history
        text = text+ ' '
        
    print(text)
    all_text.append(text)
    details['Text'] = all_text  
    text = ''
```
